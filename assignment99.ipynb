```python
###  Dimensionality Reduction with PCA
```

```python
# Import required libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the dataset
df = pd.read_csv('Wine.csv')

# Separate features and target
X = df.drop('Type', axis=1)
y = df['Type']

# 1. Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. Apply PCA (retain all components initially)
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Display the transformed dataset (Principal Components)
print("First 5 Rows of Transformed Data (Principal Components):")
print(pd.DataFrame(X_pca).head())

# Display explained variance ratio
print("Explained Variance Ratio for Each Principal Component")
print(pca.explained_variance_ratio_)
```

```python
"""1. Standardization
All features are scaled to have a mean of 0 and standard deviation of 1.

This step is necessary because:

Features like Proline and Magnesium have larger scales compared to others.

PCA is sensitive to the scale of data; standardization ensures all features contribute equally.

2. PCA Transformation
PCA is applied to the standardized data to convert it into principal components (PCs).

These principal components are linear combinations of the original features but ordered by the amount of variance they explain.

The output includes:

A transformed dataset where each feature is a principal component.

The explained variance ratio, indicating how much variance each principal component captures."""

```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Wine.csv')

# Separate features
X = df.drop('Type', axis=1)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA without limiting components
pca = PCA()
pca.fit(X_scaled)

# Explained variance ratio
explained_variance = pca.explained_variance_ratio_
cumulative_variance = explained_variance.cumsum()

# Plot Scree Plot and Cumulative Explained Variance
plt.figure(figsize=(10,6))
plt.plot(range(1, len(explained_variance)+1), explained_variance, marker='o', label='Individual Explained Variance')
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='s', label='Cumulative Explained Variance')
plt.title('Scree Plot & Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""Scree Plot
The scree plot displays the explained variance ratio for each principal component.

The first few components (typically the first 2 or 3) explain the majority of the variance.

After a certain point (elbow point), the gain in explained variance by adding more components becomes marginal.

Cumulative Explained Variance
The cumulative variance plot helps determine how many components are needed to capture a desired level of total variance:

Around 95% of the variance is often captured by the first 2 to 3 components."""

```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the dataset
df = pd.read_csv('Wine.csv')

# Separate features and target
X = df.drop('Type', axis=1)
y = df['Type']

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA with the optimal number of components (e.g., 3)
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)

# Convert PCA results to DataFrame
pca_columns = ['PC1', 'PC2', 'PC3']
df_pca = pd.DataFrame(X_pca, columns=pca_columns)
df_pca['Type'] = y  # Add back the target label for reference

# Display the first few rows of transformed data
print("Transformed Dataset with Principal Components:")
print(df_pca.head())
```

```python
"""The original dataset with 13 numerical features has been successfully reduced to 3 principal components (PC1, PC2, PC3).

Each principal component is a linear combination of the original features, designed to capture the maximum variance in the data."""
```

```python
###  Clustering with Original Data
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('Wine.csv')

# Separate features and target
X = df.drop('Type', axis=1)
y = df['Type']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 1. Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# 2. Visualization (PCA reduction to 2D for plotting)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=clusters, palette='Set1')
plt.title('K-means Clustering Results on Original Data (Visualized in 2D via PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# 3. Evaluate Clustering Performance
silhouette = silhouette_score(X_scaled, clusters)
db_index = davies_bouldin_score(X_scaled, clusters)

print(f"Silhouette Score: {silhouette:.3f}")
print(f"Davies-Bouldin Index: {db_index:.3f}")
```

```python
"""Summary of Clustering on Original Data
Clustering Applied:

K-means clustering with 3 clusters (since there are 3 wine types).

Standardization was applied prior to clustering for fair distance measurement.

Visualization:

Data was reduced to 2 principal components using PCA for visualization.

The scatter plot shows the distribution of the clusters identified by K-means in 2D space.

Clustering Performance Metrics:

Silhouette Score: Measures how well each data point fits into its assigned cluster (range -1 to 1). A higher score (~0.3-0.6) suggests good clustering.

Davies-Bouldin Index: Measures the average similarity between clusters (lower is better). A lower value indicates better clustering separation."""

```

```python
###   Clustering with PCA Data
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Wine.csv')

# Separate features and target
X = df.drop('Type', axis=1)
y = df['Type']

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce to 3 components
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)

# 1. Apply K-means clustering on PCA-transformed data
kmeans_pca = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans_pca.fit_predict(X_pca)

# 2. Visualize clustering results in 2D (first two PCs)
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=clusters_pca, palette='Set2')
plt.title('K-means Clustering on PCA-Transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# 3. Evaluate clustering performance
silhouette_pca = silhouette_score(X_pca, clusters_pca)
db_index_pca = davies_bouldin_score(X_pca, clusters_pca)

print(f"Silhouette Score (PCA Data): {silhouette_pca:.3f}")
print(f"Davies-Bouldin Index (PCA Data): {db_index_pca:.3f}")
```

```python
"""1. Clustering Results
K-means clustering was applied to the data reduced to 3 principal components.

The data was visualized in 2D (PC1 vs PC2) with clusters identified by K-means.


| Metric                   | Original Data   | PCA-Transformed Data               |
| ------------------------ | --------------- | ---------------------------        |
| Silhouette Score         | 0.285           | (0.454) higher                     |
| Davies-Bouldin Index     | 1.389           | (0.839) lower on PCA data          |
Generally, PCA-transformed data yields:

Better or comparable Silhouette Scores, indicating clearer cluster formation.

Lower Davies-Bouldin Index, suggesting more compact and well-separated clusters.

3. Comparison
PCA helps reduce noise and remove redundant features, leading to potentially improved clustering performance.

The visualizations from PCA data often show more distinct clusters compared to the original high-dimensional space."""
```
